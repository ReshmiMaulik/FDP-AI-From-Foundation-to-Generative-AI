```python
import pandas as pd
```

```python
df=pd.read_csv('/content/8 Mall_Customers (1).csv')
df
```

```python
df1=df[['Age','Annual Income (k$)','Spending Score (1-100)']]
```

```python
df1
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=5)
```

```python
kmclusters=km.fit(df1)
```

```python
df1.loc[:,'Cluster']=kmclusters.labels_

df1
```

```python
import seaborn as sns
sns.scatterplot(x='Annual Income (k$)',y='Spending Score (1-100)',data=df1,
                hue='Cluster')
```
