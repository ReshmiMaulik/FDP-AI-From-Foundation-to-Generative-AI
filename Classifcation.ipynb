```python
df1=df.dropna(axis=1)# axis 1 will delete columns
df1.shape
```

```python
df2=df.dropna(axis=0)#axis 0 will delete rows
df2.shape
```

```python
df2.info()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(max_iter=10000)
```

```python
x=df2.drop(columns='TenYearCHD')
y=df2['TenYearCHD']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

```python
lr1=LogisticRegression(max_iter=20000)
lr1.fit(x_train,y_train)
print(classification_report(y_test,lr1.predict(x_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
print(classification_report(y_test,dt.predict(x_test)))
```
